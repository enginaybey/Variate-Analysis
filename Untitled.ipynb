# Univariate,Bivariate and MultiVariate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./datasets_737503_1278636_heart.csv')
df.head()
```

```python
df.shape
```

## Univariate Analysis

```python
df_1=df.loc[df['target']==1]
df_0=df.loc[df['target']==0]
```

```python
plt.plot(df_1['age'],np.zeros_like(df_1['age']),'o')
plt.plot(df_0['age'],np.zeros_like(df_0['age']),'x')
plt.show()
```

## Bivariate Analysis

```python
sns.FacetGrid(df,hue="target",height=5).map(plt.scatter,"chol","age").add_legend();
plt.show()
```

## Multivariate Analysis

```python
sns.pairplot(df,hue="target", height=1)
plt.show()
```

```python
sns.pairplot(df, kind="reg", height=1)
plt.show()
```
